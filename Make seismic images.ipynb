```python
import numpy as np
import matplotlib.pyplot as plt
import argparse
import os

import segyio
import numpy as np
import matplotlib.pyplot as plt

from scipy import ndimage as ndi
```

```python
filename1 = 'data/Dutch Government_F3_entire_8bit seismic.segy'
seismic = segyio.tools.cube(filename1)

seismic.shape
```

```python
def make_image(arr,outputname):
    fig = plt.imshow(arr)
    fig.set_cmap('seismic')
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    plt.savefig('images/' + outputname)
    
#make_image(seismic[:,200,:].T,f'seismicxline{200}')
```

```python
for i in range(0,seismic.shape[0]):
    make_image(seismic[i,:,:].T,f'seismicinline{i}')
    
```

```python
for i in range(0,seismic.shape[1]):
    make_image(seismic[:,i,:].T,f'seismicxline{i}')
```
